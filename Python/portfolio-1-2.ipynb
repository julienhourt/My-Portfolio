Everything is okay for the previous columns, I have decided to keep history as it is and to not transform cases like for example "Germany DR" and "Germany" or for "Zaire" wich is the previous name of "Congo Republic".

```python
#Round column

#Checking how values are organized in "Round"
values_round = df.Round.unique()
print(values_round)
```

```python
#We have to harmonize it

df.loc[df['Round'] == 'Final stage', 'Round'] = 'Final'
df.loc[df['Round'] == 'First round', 'Round'] = 'Group stage'
df.loc[df['Round'] == 'Second round', 'Round'] = 'Group stage'
df.loc[df['Round'] == 'First group stage', 'Round'] = 'Group stage'
df.loc[df['Round'] == 'Second group stage', 'Round'] = 'Group stage'
df.loc[df['Round'] == 'Group stage play-off', 'Round'] = 'Group stage'

#Checking 
new_values_round = df.Round.unique()
print(new_values_round)
```
